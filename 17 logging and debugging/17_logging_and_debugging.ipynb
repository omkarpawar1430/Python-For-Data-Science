### debugging

identify and fix errors

```python
# simple / root logger

import logging

logging.basicConfig(filename="txt.log",  
                    level = logging.DEBUG,
                    filemode="a", 
                    format="%(asctime)s : %(levelname)s : %(message)s")

logger = logging.getLogger()

logger.setLevel(logging.WARNING)

try:
    logger.info("program initiated...")
    x = 12 / 0
    logger.debug(f"result is {x}")
except Exception as e:
    logger.error(f"Error: {e} ocurred.")
else:
    logger.info("message written successfully...")
finally:
    logger.info("sub to data decides")
```

```python
import logging

logger = logging.getLogger("my_logger")
logger.setLevel(logging.DEBUG)

format = logging.Formatter("%(asctime)s : %(name)s : %(levelname)s : %(message)s")

# debug and info:
my_file_handler = logging.FileHandler("debug_info.log")
my_file_handler.setLevel(logging.DEBUG)
my_file_handler.setFormatter(format)

# warning and error:

console_handler = logging.FileHandler("warning_error.log")
console_handler.setLevel(logging.WARNING)
console_handler.setFormatter(format)

logger.addHandler(my_file_handler)
logger.addHandler(console_handler)


try:
    logger.info("program initiated...")
    x = 12 / 0
    logger.debug(f"result is {x}")
except Exception as e:
    logger.error(f"Error: {e} ocurred.")
else:
    logger.info("message written successfully...")
finally:
    logger.info("sub to data decides")
```

```python
import logging

logger = logging.getLogger("my_logger")
logger.setLevel(logging.DEBUG)

format = logging.Formatter("%(asctime)s : %(name)s : %(levelname)s : %(message)s")

# logging:
my_file_handler = logging.FileHandler("my_file.log")
my_file_handler.setLevel(logging.DEBUG)
my_file_handler.setFormatter(format)

# printing in console:

console_handler = logging.StreamHandler()
console_handler.setFormatter(format)

logger.addHandler(my_file_handler)
logger.addHandler(console_handler)


try:
    logger.info("program initiated...")
    x = 12 / 0
    logger.debug(f"result is {x}")
except Exception as e:
    logger.error(f"Error: {e} ocurred.")
else:
    logger.info("message written successfully...")
finally:
    logger.info("sub to data decides")
```
