# For Loop | Python for Data Science

### Syntax:

#### Using with `range()`:

```python
list(range(10)) # start # end ( which not included) # step
```

```python
for i in range(0, 10, 2):
    print(i)
```

#### Using with `iterables`:

```python
shopping_list = {"kindle":9999, "Naruto T shirt": 236, "Camera": 45999}
bill = 0
for i in shopping_list:
    bill = bill + shopping_list[i]
print(bill)
```
